## Data Exploration

Select data and take a look of the dataset

```python
%%sql

SELECT *
FROM covid_deaths
LIMIT 1000;
```

As we can see, We have tons of data and many columns. We are not gonna work on all of them but few. Lets select toal_cases, new_cases, total_deaths and poplution. 

```python
%%sql

SELECT location, date, total_cases, new_cases, total_deaths, population
FROM covid_deaths
ORDER BY 2
LIMIT 100;
```

BY running the query above we can clearly see that the first ever infected cases and deaths were spoted in china on January 22, 2020. Total death counted that day is 17 and total cases were 548. 

Lets find out the overall death ratio compared to total cases from my home country which is Bangladesh. 

```python
%%sql 

-- Total Cases vs Total Deaths

SELECT location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 AS Death_percentage
FROM covid_deaths
WHERE location like '%desh%' and continent is not null
ORDER BY 5 DESC, 2
LIMIT 100;
```

Now we gonna look at the total infected ratio compared to overall population in Bangladesh.  

```python
%%sql 

SELECT location, date, total_cases, population, (total_cases/population)*100 AS cases_percentage
FROM covid_deaths
WHERE location like '%desh%' and continent is not null
ORDER BY 5 DESC, 2
LIMIT 100;
```

In above queries we used the condition "Continenet is not null" in WHERE statetement is because, If we look at the dataset closely we find that the location column contain 'Africa', 'Asia', 'Europe' as a country. Only when the continent is null. To be more specific and acurate we used "continent is not null" in the WHERE statement.

Lets have look!

```python
%%sql 

SELECT *
FROM covid_deaths
WHERE continent is null
LIMIT 100;
```

Now, Lets go global and find those countries has a most highest infection rate compared to their population

```python
%%sql 

SELECT location, population, MAX(total_cases) AS highestCases, MAX((total_cases/population)*100) AS highestInfectedRate
FROM covid_deaths
WHERE continent is not null
GROUP BY location, population
ORDER BY 4 DESC;
```

As we see, Andorra, Montenegro and Czechia has the highest infected cases among 210 countries from world.  

Now lets see which countries had the most death. 

```python
%%sql

Select Location, max(total_deaths) as highestDeathCount
From covid_deaths
Where continent is not null
Group by Location
Order by 2 DESC;
```
